```python
%matplotlib inline
import os
import glob
import numpy as np
import email
import pandas as pd
```

Get emails path by using the glob module which is used to retrieve files/pathnames matching a specified pattern

```python
easy_ham_paths = glob.glob('easy_ham/*')
easy_ham_2_paths = glob.glob('easy_ham_2/*')
hard_ham_paths = glob.glob('hard_ham/*')
spam_paths = glob.glob('spam/*')
spam_2_paths = glob.glob('spam_2/*')
```

```python
len(easy_ham_paths)
```

Get email Content Function by using email module .

```python
def get_email_content(email_path):
    file = open(email_path,encoding='latin1')
    try:
        msg = email.message_from_file(file)
        for part in msg.walk():
            if part.get_content_type() == 'text/plain':
                return part.get_payload() # prints the raw text
    except Exception as e:
        print(e)
```

```python
def get_email_content_bulk(email_paths):
    email_contents = [get_email_content(o) for o in email_paths]
    return email_contents
```

An Example of content of a ham email 

```python

print(get_email_content(easy_ham_paths[0]))
```

```python
ham_path = np.array([
    easy_ham_paths,
    easy_ham_2_paths,
    hard_ham_paths
],dtype=object)

spam_path = np.array([
    spam_paths,
    spam_2_paths
],dtype=object)
```

# Ham Data

```python
ham_sample = np.array([get_email_content_bulk(o) for o in ham_path])
```

```python
ham_sample = np.concatenate((ham_sample[:]))
```

```python
def remove_null(data):
    
    return np.array([i for i in data if i is not None])
```

```python
ham_sample.shape
```

```python

ham_sample = remove_null(ham_sample)
```

```python
ham_sample.shape
```

```python
#negative class
zeros = np.zeros((ham_sample.shape[0],))
```

```python
zeros.shape
```

```python
ham_sample = np.vstack((zeros,ham_sample))
```

```python
ham_sample = ham_sample.transpose()
```

```python
DF = pd.DataFrame(ham_sample)
  
# save the dataframe as a csv file
DF.to_csv("DirtyHam.csv")
```
